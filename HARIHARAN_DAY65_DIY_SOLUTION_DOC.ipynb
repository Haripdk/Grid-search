```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('voice.csv')
df.head()
```

```python
df.rename(columns = {'label':'Gender_Identified'}, inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df.Gender_Identified = le.fit_transform(df['Gender_Identified'])
```

```python
X=df.drop(['Gender_Identified'],axis=1)
y=df.Gender_Identified
```

```python
X
```

```python
y
```

```python
params = { "criterion" : ["gini", "entropy"], "n_estimators": [100, 150, 200,300]}
rf_gsv=GridSearchCV(estimator=RandomForestClassifier(),param_grid=params,cv=3,scoring='accuracy')
rf_gsv.fit(X,y)
```

```python
pd.DataFrame(rf_gsv.cv_results_).sort_values('rank_test_score')
```

```python
print("The best parameters are:")
rf_gsv.best_params_
```

```python
print("The best estimator is:")
rf_gsv.best_estimator_
```
